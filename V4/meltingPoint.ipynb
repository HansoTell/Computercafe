```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv("Data1.csv", decimal = ",")
db1 = pd.read_csv("Gefrierpunkterniedrigung1.csv", decimal = ",")

db.head()
```

```python
fig, ax = plt.subplots()
ax.plot(db["Wert1"], db["Wert4"])
plt.xlabel("Zeit [s]", size=14)
plt.ylabel("Temperatur [C]", size=14)
plt.grid()
plt.show()
```

```python
plt.plot(db1["Zeit"], db1["Temperatur"], color = "black")
plt.xlabel("Zeit [s]", size=14)
plt.ylabel("Temperatur [C]", size=14)
plt.grid(True)
plt.show()
```

```python
index = db1["Temperatur"].idxmin()
db_extrapol = db1.loc[index:]
db_extrapol.head()
```

```python
m,b = np.polyfit(db_extrapol["Zeit"], db_extrapol["Temperatur"], 1)
print(f"y-Abschnitt: {b:.4f}")


```
