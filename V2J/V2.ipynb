## Theoretische Grundlagen
### Wellen

Wellen sind periodische Störungen in Raum und Zeit. Sie breiten sich allgemein
dreidimensional im Raum aus. Wenn ein Stein  in einen Teich  geworfen wird, entstehen
konzentrische Kreise, die mit der Zeit immer weiter von der Eintauchstelle weglaufen.
Würde man den Querschnitt der Wasseroberfläche fotografieren, bilden die
unterschiedlichen Wasserhöhen ein räumliches periodisches Muster zu einer festen Zeit.
Den Abstand zwischen zwei benachbarten Maxima nennt man **Wellenlänge** $\lambda$.
Andererseits könnte man die Wasserhöhe an genau einem festen Punkt filmen und dabei
sehen, wie die Wassergrenzfläche an diesem Punkt sich periodisch in der Zeit auf und ab
bewegt. Die Zeit zwischen zwei aufeinanderfolgenden Maxima nennt man die Periodendauer T. 
Das reziproke der Periodendauer gibt an, wie oft pro Sekunde eine vollständige zeitliche Periode durchlaufen wird und wird **Frequenz** $\nu = 1/T$ genannt.  

Mathematisch wird eine Welle allgemein durch folgende Funktion beschrieben:
$$
    y_1 = A \cdot \sin (\vec{k} \cdot \vec{r} - \omega t) 
$$

Dabei bezeichnet $y_1$ den Bertrag der Wellenfunktion am Ort $\vec{r}$ zur Zeit t,
A die (maximale) Amplitude der Welle, $\vec{k}$ den Wellenvektor und $\omega$ die Kreisfrequenz. Der Wellenvektor gibt
die Ausbreitungsrichtung der Welle an. Dessen Betrag (genannt „Wellenzahl“) k ist folgendermaßen mit der Wellenlänge $\lambda$ verknüpft
$$
    k = \frac{2\pi}{\lambda} 
$$
Diese Zahl gibt an, um wieviel der räumliche Phasenwinkel pro Längeneinheit
voranschreitet (in der Einheit "rad": $360\degree$ entsprechen $2\pi$ im Bogenmaß (rad)).
Die Kreisfrequenz
$$
    \omega = \frac{2\pi}{T} = 2\pi \nu
$$
gibt an, wie weit der zeitliche Phasenwinkel (in rad) pro Zeiteinheit voranschreitet.

Häufig reicht für die prinzipielle Beschreibung von Wellenphänomenen auch eine
räumlich eindimensionale Darstellung:
$$
    y_1 = A \cdot \sin (kx - \omega t)
$$
Mit  dieser  Gleichung kann auch eine elektromagnetische Welle („Licht“) beschrieben
werden: Dann ist A der Betrag oder die Amplitude des elektrischen Feldstärkevektors
(der steht i.A. immer senkrecht zur Ausbreitungsrichtung, die magnetische Komponente
wird hier nicht weiter berücksichtigt).

momentane  Auslenkung y  in  Abhängigkeit  vom  Ort x  aufgetragen.  Die  Wellenberge
bewegen sich mit der Phasengeschwindigkeit $\vec{v_p}$ durch den Raum. Für Licht im Vakuum
hat die Phasengeschwindigkeit den Wert der Lichtgeschwindigkeit $c_0$ܿ

### Interferenz
Das Überlagern von zwei oder mehr Wellenamplituden nennt man Interferenz. Trifft ein
Wellenmaximum einer Welle auf ein Wellenminimum einer anderen, kommt es zu einer
Auslöschung  der  Störung  (wenn  beide  Amplituden  gleich  groß  sind);  dieses  Phänomen
heißt  „destruktive  Interferenz“.  Treffen  sich  dagegen  zwei  Maxima,  kommt  es  zu  einer
Verstärkung der Störung und wird „konstruktive Interferenz“ genannt.


<div style="text-align:center;">
    <img src="ueberlagerung_wellen.png" width=600>
    <p><b>Abbildung 2.2:</b> Überlagerung zweier Sinus-Wellen mit Gangunterschied </p>
</div>


Mathematisch wird das einfach durch Addition von (zwei) Sinus-Funktionen beschrieben (siehe Abbildung 2.2); Eine der beiden Wellen hat gegenüber der anderen einen Gangunterschied $\delta$.
$$
    \begin{align*}
        y_1 &= A \cdot \sin (kx) \\
        y_2 &= A \cdot \sin (k(x + \delta ))
    \end{align*}
$$
Für  die  Überlagerung  von  Lichtwellen  spielt  die  Zeitabhängigkeit  in  erster  Linie  keine
Rolle und wurde daher weggelassen.  

Die Überlagerung der beiden Wellen ergibt
$$
    y = y_1 + y_2 = A \cdot (\sin (kx) + \sin(l(x + \delta )))
$$
Die Summe auf der rechten Seite kann maximal werden, wenn z.B. $\delta = 0$ ist. Wegen der
Periodizität ist die konstruktive Interferenz auch möglich, wenn der Gangunterschied ein
**ganzzahliges Vielfaches der Wellenlänge** ist, insgesamt also:
$$
    \delta = n \cdot \lambda \quad n = 0,1,2,...
$$
Destruktive  Interferenz  tritt  auf,  wenn  die  Summe  auf  der  rechten  Seite  0  wird.  Das
geschieht, wenn der Gangunterschied in **ungeradzahliges Vielfaches der halben
Wellenlänge** ist:
$$
    \delta = (2n +1) \cdot \frac{\lambda}{2} \quad n=0,1,2,..
$$

### Beugung
Unter  Beugung  versteht  man  die  Ablenkung  von  Wellen,  die  durch  Interferenz  der
Teilwellenuntereinander  entsteht.  Beugung  beobachtet  man,  wenn  eine  Welle  auf  ein
Hindernis  trifft,  zum  Beispiel  einen  Spalt,  ein  Gitter,  eine  Lochblende  oder  ähnliches.  Je
nach Beschaffenheit des Hindernisses entsteht dabei ein charakteristisches
Beugungsmuster. Beugung wird durch das Huygens‘sche Prinzip erklärt: jeder Punkt des
Hindernisses wird Ausgangspunkt von neuen Kugelwellen. Dadurch kommt es zu
Überlagerung von Kugelwellen:

<div style="text-align:center;">
    <img src="Kugelwellen.png" width=600>
    <p><b>Abbildung 2.5:</b> Entstehung der Beugung durch Überlagerung von Kugelwellen a) Eine Kugelwelle. b) 2 Kugelwellen etwas verschoben c) 3 Kugelwellen etwas verschoben</p>
</div>

In Abbildung 2.5 ist der Effekt der Überlagerung mehrerer Kugelwellen dargestellt: in a)
sind die Maxima einer einzelnen Kugelwelle als konzentrische Kreise abgebildet. In b) ist
die  Überlagerung  von  2  solchen  Kugelwellen  gezeichnet.  Immer,  wenn  zwei  Linien  sich
exakt überdecken, entsteht konstruktive Interferenz; wo Linien „nebeneinander“ sind, ist
die  Interferenz  partiell  destruktiv.  Man  erkennt,  dass  bereits  bei  2  Kugelwellen  die
konstruktive Interferenz nur in bestimmte Richtungen erfolgt, symbolisch durch 3
beispielhafte Pfeile dargestellt. In Bild c) sind drei Kugelwellen überlagert. Die Richtungen
für konstruktive Interferenz bleiben im wesentlichen dieselben, sind aber von der
Winkelbreite deutlich schmaler, weswegen die Pfeile auch schmaler dargestellt sind.  

Eine Realisation  dieser  Beugung ist die Durchstrahlung eines Strichgitters mit Licht wie
in Abbildung 2.6 dargestellt.  Der Abstand zwischen 2 Strichen heißt **Gitterkonstante** d:
Jede  Öffnung  ist  nun  Ausgangspunkt  neuer  Kugelwellen.  Man  zeichnet  aber  nicht  mehr
die konzentrischen Kreise, sondern nur noch die konstruktiv interferierten Teilstrahlen.
Zwei benachbarte Strahlen durchlaufen unterschiedliche Weglängen bis zu einem Schirm;
dieses nennt man Gangunterschied. Der Gangunterschied ist in der Skizze mit $\delta$ bezeichnet.  
Die  Bedingung  für  konstruktive  Interferenz  ist,  dass  der  Gangunterschied
gerade ein ganzzahliges Vielfaches der Wellenlänge beträgt, d.h., es muss gelten:
$$
    d \cdot \sin \beta_n = n \cdot \lambda
$$
Die Zahl n deutet an, dass es mehrere Richtungen konstruktiver Interferenz gibt.

Bei  bekannter  Gitterkonstante d  kann  die  Wellenlänge  einer  unbekannten  Strahlung
bestimmt werden. Kennt man die Wellenlänge, kann die Gitterkonstante ermittelt
werden. Das ist auch das Prinzip der Röntgenstrukturuntersuchung.

### Bekanntes Gitter / unbekannter Laser: Auswertung der Wellenlänge nach einzelnen Reflexen und Mittelwertbildung

Aus den so ermittelten Parametern $z_0$ und $y_n$ werden mithilfe der Gleichungen die Winkel $\beta_{-n}$ und $\beta_n$ für die berechnet. Einsetzen in die Gleichung erlaubt die Berechnung der Wellenlänge für jeden der Streuwinkel. Der Mittelwert dieser sechs Werte ergibt die mittlere Laserwellenlänge $\lambda_{mittel}$

### Bekanntes Gitter / unbekannter Laser: Auswertung der Wellenlänge nach einzelnen Reflexen gleichzeitig aus einer Auftragung

Anschließend wird die Laserwellenlänge durch eine Auftragung derselben Daten
bestimmt. Gleichung (2.10) lässt sich zu
$$
    \sin \beta_n = \frac{n}{d} \cdot \lambda
$$
umstellen. Aus der Auftragung von $\sin \beta_n$ vs. n/d erhält man die Laserwellenlänge $\lambda$ als Steigung.

### Unbekanntes Gitter / bekannter Laser: Auswertung der Gitterkonstante nach einzelnen Reflexen und Mittelwertbildung

Die  Gitterkonstante d  eines  unbekannten  Gitters  lässt  sich  berechnen,  indem  Gleichung
(2.10) nach d aufgelöst wird:
$$
    d = \frac{n \cdot \lambda}{\sin \beta_n}
$$

Aus den Werten für d lässt sich ein Mittelwert $d_{mittel}$ berechnen

### Unbekanntes Gitter / bekannter Laser: Auswertung der Gitterkonstante durch alle Reflexe gleichzeitig

Auch d lässt sich aus einer Auftragung ermitteln. Dazu wird $n \cdot \lambda$ vs. $\sin \beta_n$ aufgetragen, woraus man d aus der Steigung erhält.
