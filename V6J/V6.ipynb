## 1 Theoretische Grundlagen

Der stoffliche, chemische Gleichgewichtszustand für ein isoliertes homogenes System mit
den Komponenten i ist der Zustand, der sich nach unendlich langer Zeit (oder: nach allen
denkbaren  Störungen)  mit  den  Gleichgewichtskonzentrationen $c_i$ eingestellt hat. Die
Gleichgewichtskonstante K, für ein chemisches Gleichgewicht mit den stöchiometrischen
Koeffizienten $v_i$ (negativ für Edukte, positiv für Produkte)
$$
    |v_A |A + |v_B |B \;\rightleftharpoons\; v_cC + v_DD \tag{1.1}
$$

ist nach dem Massenwirkungsgesetzt (MWG) definiert durch:

$$
    K = c_C^{v_c} \cdot c_D^{v_D}\cdot c_A^{v_A} \cdot c_B^{v_B} \cdot (c^0)^{\sum v_i} = \frac{c_C^{v_c} \cdot c_D^{v_D} }{c_A^{|v_A|} \cdot c_B^{|v_B|}} \cdot (c^0)^{\sum v_i} \tag{1.2}
$$

Einbeziehung der Standardkonzentration $c^0 = 1 \; mol \; dm^{-3}$ bewirkt, dass die
Gleichgewichtskonstante stets dimensionslos ist. Die Gleichgewichtskonstante K variiert
mit den äußeren Bedingungen wie Temperatur, Druck und Lösungsmittel.

### 1.1 Säure-Base-Gleichgewicht
Stellt sich für ein Stoffgemisch, wie die schwache Säure HR in Wasser, in homogener Phase
das chemische Gleichgewicht

$$
    HA \;\rightleftharpoons\; H^+ + A^- \tag{1.3}
$$

ein, so gilt für die Konzentrationen c im Gleichgewicht das Massenwirkungsgesetz:

$$
    K_S = \frac{c(H^+ \cdot c(A^-))}{c(HA)} \cdot \frac {1}{c^0} \tag{1.4}
$$

Die Gleichgewichtskonstante $K_S$ ist eine Funktion von Temperatur und Druck. Bei Säure-
(s engl. acid: a) bzw. Basegleichgewichten (b; engl. base, b) wird häufig auch die
Bezeichnung $K_a$ bzw. $K_b$ verwedet.  

Der pH-Wert einer Lösung mit der Wasserstoffionenkonzentration c($H^+$) ist definiert als:

$$
    pH = -\log_{10} (\frac{c(H^+)}{c^0}) \tag{1.5}
$$

Der $pK_S$-Wert einer  Säure  mit der  Gleichgewichtskonstanten $K_S$ für das  chemische
Gleichgewicht der Säuredissoziation ist analog definiert als:

$$
    pK_S = -\log_{10} K_S \tag{1.6}
$$

Bei wässrigen Lösungen ist zusätzlich die Eigendissoziation des Wassers zu
berücksichtigen. Für dieses Gleichgewicht

$$
    H_2O \;\rightleftharpoons\; H^+ + OH^- \tag{1.7}
$$

gilt das Massenwirkungsgesetz bei $T=25 \degree C$

$$
    K_W = \frac {c(H^+) \cdot c(OH^-)}{(c^0)^2} = 10^{-14} \tag{1.8}
$$

$K_W$ wird als **Ionenprdoukt** des Wassers bezeichnet.  

Im  Folgenden  soll  die  Berechnung  des  pH-Wertes  für  verschiedene  Systeme,  in  denen
Säure-Base-Gleichgewichte eine Rolle spielen, erfolgen. Die Herleitung der notwendigen
Gleichungen  erfolgt  über  den  Bilanzgleichungsansatz.  Folgende  Bilanzgleichungen  bzw.
Massenwirkungsgesetze müssen daher jeweils berücksichtigt werden:

1. Massenerhaltung, Stoffbilanz, Elementerhaltung
2. Elektroneutralität
3. Massenwirkungsgesetz für Wasserdissoziation
4. Massenwirkungsgesetz für die Säure / Base

**Schwache Säuren**  

Das Säuregleichgewicht wird nach folgendem Schema berechnet:
| Nr | Art | Formel | Referenz |
|----:|-----------:|:--------:|----------:|
| 1. | Einwaage der schwachen Säure HA in Wasser | $c_1$ | (1.9) |
| 2. | Anzahl beteiligter Teilchensorten: | $HA, A^-, H^+, OH^-$ | | 
| 3. | Stoffmengenbilanz | $c(HA) + c(A^-) = c_1$ | (1.10) |
| 4. | Elektroneutralität | $c(H^+) = c(OH^-) + c(A^-)$ | (1.11) |
| 5. | Ionenprodukt des Wassers | $K_W = \frac{c(H^+) \cdot c(A^-)}{(c^0)^2}$ | (1.12) |
| 6. | Massenwirkungsgesetz der Säure | $K_S = \frac{c(H^+) \cdot c(A^-)}{c(HA)} \cdot \frac{1}{c^0}$ | (1.13) |

Die  Anzahl  der  Gleichungen  (4)  reicht  aus,  um  die  4  unbekannten  Konzentrationen  bei
Kenntnis  von $c_1, K_W$ und $K_S$ auszurechnen.  Durch  Einsetzen  der  Gleichungen (1.10)  bis
(1.12) in Gleichung (1.13) erhält man einen Ausdruck für die Säurekonstante $K_S$, in dem nur noch bekannte Größen auftreten:

$$
    K_S = \frac{c(H^+)}{c^0} \cdot \frac{c(H^+ - c(OH^-))}{c_1 - (c(H^+) - c(OH^-))} = \frac{c(H^+)}{c^0} \cdot \frac{c(H^+) - \frac{K_W (c^0)^2}{c(H^+)}}{c_1 - (c(H^+) - (\frac{K_W (c^0)^2}{c(H^+)}))} \tag{1.14}
$$

Die  implizite  Gleichung  (1.14)  lässt  sich  analytisch  lösen;  jedoch  ist  das  kompliziert.
Geeignete  Näherungen  vereinfachen  die  Lösung.  Im  sauren  pH-Bereich  ist $c(H^+ ) \gg c(OH^-)$:

$$
    K_S = \frac{(c(H^+)^2)}{(c1 - c(H^+))c^0} \rightarrow c(H^+) = -\frac{K_S \cdot c^0}{2} + \sqrt{(\frac{K_S \cdot c^0}{2})^2 + K_S c_1 c^0} \tag{1.15}
$$

Aus der Näherung $c_1 \gg c(H^+)$ (schwache Säure) folgt weiter:

$$
    c(H^+) = \sqrt{K_S \cdot c_1 \cdot c^0} \tag{1.16}
$$

Und mit der Definition des pH-Wertes gilt dann:

$$
    pH = \frac{1}{2} \cdot (pK_S - \log_{10}(\frac{c_1}{c^0})) \tag{1.17}
$$  

**Pufferlösung**  

Als  Pufferlösung  wird  eine  Lösung  einer  schwachen  Säure  HR  (Einwaagekonzentration $c_1$) und ihres Salzes MR (Einwaagekonzentration $c_2$) bezeichnet, da bei der Zugabe einer
kleinen Menge starker Säure HX (Einwaagekonzentration $c_3$) der pH-Wert der Lösung (in
gewissen Grenzen) konstant gehalten wird. Nach dem allgemeinen Schema gelten für eine
Pufferlösung ohne Zugabe einer starken Säure folgende Bilanzgleichungen bzw.
Massenwirkungsgesetze:

| Nr | Art | Formel | Referenz |
|----:|-----------:|:--------:|----------:|
| 1. | Einwaage der schwachen Säure HA in Wasser <br> Einwaage ihres Salzes | $c_1$ <br> $c(M^+) = c_2$ | (1.18) |
| 2. | Anzahl beteiligter Teilchensirten: | $HA, A^-, H^+, OH^-, M^+$ | |
| 3. | Stoffmengenbilanz | $c(HA) + c(A^-) = c_1 + c_2$ | (1.19) |
| 4. | Elektroneutralität | $c(H^+) + c(M^+) = c(OH^-) + c(A^-)$ | (1.20) |
| 5. | Ionenprodukt des Wassers | $K_W = \frac{c(H^+) \cdot c(A^-)}{(c^0)^2}$ | (1.21) |
| 6. | Massenwirkungsgesetz der Säure | $K_S = \frac{c(H^+) \cdot c(A^-)}{c(HA)} \cdot \frac{1}{c^0}$ | (1.22) |

Durch  Umstellen  der  Gleichungen (1.18)  bis (1.21)  und  Einsetzen  in  Gleichung (1.22)  bis
erhält man folgenden vereinfachten Ausdruck:

$$
    K_S = \frac{c(H^+)}{c^0} \cdot \frac{c_2 + c(H^+) - c(OH^-)}{c_1 - (c(H^+) - c(OH^-))} \tag{1.23}
$$

Mit $c_1, c_2 \gg |c(H^+) - c(OH^-)|$ gilt näherungsweise:

$$
    K_S = \frac{c(H^+)}{c^0} \cdot \frac{c_2}{c_1} \tag{1.24}
$$

Damit gilt für den pH-Wert:

$$
    pH = pK_S + \log_{10} (\frac{c_2}{c_1}) \tag{1.25}
$$

Mit  dieser  als  Henderson-Hasselbalch-Gleichung  bekannten  Formel  lässt  sich  der  pH-
Wert einer Pufferlösung berechnen. Für den Fall $c_1=c_2$ folgt $pH = pK_S$.  


**Pufferlösung und Säure**  

Bei Säurezugabe wird der pH-Wert nahezu konstant gehalten und die Pufferung ein wenig
verbraucht.  Die  Auswirkung  der  Säurezugabe  auf  die  im  Gleichgewicht  befindlichen
Konzentrationen  der  einzelnen  Komponenten  der  Pufferlösung  lässt  sich  nach  dem
allgemeinen,  oben  bereits  angewandten  Schema  ableiten.  Nach  Zugabe  einer  starken
Säure mit der Einwaagekonzentration $c_3$ zu der Pufferlösung gelten folgende
Bilanzgleichungen und Massenwirkungsgesetze:

| Nr | Art | Formel | Referenz |
|----:|-----------:|:--------:|----------:|
| 1. | Einwaage der schwachen Säure HA in Wasser <br> Einwaage ihres Salzes <br> Einwaage der starken Säure | $c_1$ <br> $c(M^+) = c_2$ <br> $c(X^-) = c_3$ | (1.26) |
| 2. | Anzahl beteiligter Teilchensorten: | $HA, A^-, H^+, OH^-, M^+, X^-$ | |
| 3. | Stoffmengenbilanz | $c(HA) + c(A^-) = c_1 + c_2$ | (1.27) |
| 4. | Elektroneutralität | $c(H^+) + c(M^+) = c(OH^-) + c(A^-) + c(X^-)$ | (1.28) | 
| 5. | Ionenprodukt des Wassers | $K_W = \frac{c(H^+) \cdot c(A^-)}{(c^0)^2}$ | (1.29) |
| 6. | Massenwirkungsgesetz der Säure | $K_S = \frac{c(H^+) \cdot c(A^-)}{(HA)} \cdot \frac{1}{c^0}$ | (1.30) | 

Für die Berechnung der sechs unbekannten Konzentrationen stehen sechs unabhängige
Gleichungen zu Verfügung. Analog nach den bisherigen Verfahren erhält man den
folgenden vereinfachten Ausdruck für die Säurekonstante $K_S$:

$$
    K_S = \frac{c(H^+)}{c^0} \cdot \frac{c_2 - c_3 + c(H^+) - c(OH^-)}{c_1 + c_2 - (c_2 - c_3 + c(H^+) - c(OH^-))} \tag{1.31}
$$

Von praktischem Interesse ist der Fall mit $c_1, c_2 \gg c_3$ und $(c_2-c_3) \gg |c(H^+) - c(OH^-)|$.  
Dann gilt näherungsweise:

$$
    K_S = \frac{c(H^+)}{c^0} \cdot \frac{c_2 - c_3}{c_1 + c_3} \tag{1.32}
$$

Damit gilt für den pH-Wert

$$
    pH = pK_S + \log_{10} (\frac{c_2 - c_3}{c_1 + c_3}) \tag{1.33}
$$

Die  letzte  Gleichung  ähnelt  der  Henderson-Hasselbalch-Gleichung.  Mit  ihr  lässt  sich  die
Änderung  des  pH-Wertes  nach  Zugabe  kleiner  Mengen  einer  starken  Säure  zu  einer
Pufferlösung berechnen.
