## 1 Theoretische Grundlagen

Eine **Phase** ist ein homogener Bereich der Materie von makroskopischer Größe.  

Der **Sättigungsdampfdruck** ist derjenige Druck, der sich für die koexistierenden Phasen
fest bzw. flüssig und gasförmig bei vorgegebener Temperatur im Gleichgewicht einstellt.  

Die **normale  Schmelztemperatur**  (bzw.  Siedetemperatur)  ist  die  Temperatur,  bei  der
ein  Phasengleichgewicht  flüssig/fest  (bzw.  gasförmig/flüssig)  bei $p = 101 325$ Pa (= 1 atm) vorliegt

### 1.1 Dampfdrücke binärer Lösungen

Der  Dampfdruck  einer  idealen  binären  Lösung  setzt  sich  additiv  aus  den  partiellen
Dampfdrücken der einzelnen Komponenten zusammen. Unter „ideal“ versteht man, dass
die Wechselwirkungen zwischen allen Molekülen identisch sind, egal von welcher
Komponente  sie  stammen.  Für  den  partiellen  Dampfdruck  einer  Komponente i  der
idealen Mischung gilt bei einer vorgegebenen Temperatur T das Raoultsche Gesetz:
$$
    p_i = p_{0i} \cdot x_i^l \tag{1.1}
$$

Dabei ist $p_{0i}$ der Dampfdruck der reinen Komponente i und $p_i$  ist  der  Dampfdruck  der
Komponente i in einer Mischung mit dem Stoffmengenanteil $x_i^l$ in der flüssigen Phase (l) 
Das Raoultsche Gesetz gilt für ideale Mischungen im gesamten Bereich des
Stoffmengenanteils $x_i^l$.  

Für  reale  Mischungen  zweier  flüssiger  Komponenten  gilt  für  den  Partialdruck  einer
Hauptkomponente  1  (Lösungsmittel)  das  Raoultsche  Gesetz  nur  noch  im  Grenzbereich
einer stark verdünnten Lösung, also für $x_2^l << 1$. Aufgrund von Wechselwirkungen
zwischen  den  Molekülen  kommt  es  bei  größeren  Stoffmengenanteilen  des  Gelösten  zu
positiven oder negativen Abweichungen vom Raoultschen Gesetz.  

In realen Lösungen, bei denen der Dampfdruck der gelösten Komponente annähernd null
ist ($p_2  \approx 0$, z.B bei Feststoffen), gilt das  Raoultsche Gesetz  bei kleinen Konzentrationen
des Gelösten. Setzt man in Gleichung (1.1) für $x_1^l = (1-x_2^l)$ ein, so erhält man: 
$$
    \begin{align*}
        p_1 &= p_{01} - p_{01} \cdot x_2^l \\ \tag{1.2}
        p_{01} - p_1 &= \Delta p = p_{01} \cdot x_2^l
    \end{align*}
$$

Man kann in diesem Fall das Raoultsche Gesetz daher auch wie folgt formulieren: In einer
verdünnten Lösung ist die relative Dampfdruckerniedrigung des Lösungsmittels
proportional zum Stoffmengenanteil des gelösten Stoffes in der Lösung.  

Die  Aussage,  der  Dampfdruck  über  einer  verdünnten  Lösung  sei  stets  kleiner  als  der
Dampfdruck des reinen Lösungsmittels, ist also in dieser Form falsch. Sie gilt nur, wenn
der  Partialdruck  des  gelösten  Stoffes  so  klein  ist,  dass  der  Gesamtdampfdruck  über der
Lösung praktisch gleich dem Partialdampfdruck des Lösungsmittels ist.  

Verglichen mit dem idealen Gasgesetz ist es für das Raoultsche Gesetz schwerer
anzugeben,  bis  zu  welchen  Konzentrationen  es  eine  gute  Näherung  darstellt.  Es  gibt
einzelne  Mischungen,  in  denen  es  im  ganzen  Mischungsbereich  annähernd  gilt  und
andere, bei denen bereits für $x_2^l < 0.01$ deutliche Abweichungen  nachweisbar sind. Alle
einfachen Beziehungen für verdünnte Lösungen gelten jedenfalls nur in dem
Konzentrationsbereich, in dem auch Gleichung (1.1) gilt.

### 1.2 Phasendiagramm und Gefrierpunktserniedrigung 

Für  eine  reine  Substanz  hängt  der  Dampfdruck  bzw.  der  Schmelzdruck  nur  von  der
Temperatur  der  Flüssigkeit  ab.  Dagegen  werden  für  ein  binäres  Gemisch  sowohl  der
Dampfdruck  als  auch  der  Schmelzdruck  zusäzlich  von  der  Konzentration  des  Gelösten
abhängen.  Wir  wollen  im  folgenden  nur  hochverdünnte  binäre  Gemische  betrachten,
deren Gasphase und feste Phase jeweils nur aus dem reinen Lösungsmittel (Substanz 1)
bestehen. Das heißt:  

- Der Dampfdruck des Gelösten (Substanz 2) ist vernachlässigbar klein.
- Beim Kristallisieren entsteht nur reine Lösungsmittelkristalle, keine Mischkristalle

Die  Unterschiede  zwischen  der  Dampfdruckkurve  bzw.  der  Schmelzdruckkurve  einer
solchen Lösung und des reinen Lösungsmittels lassen sich anschaulich in einem
Phasendiagramm  (p-T-Diagramm)  darstellen.  In  unserem  Fall  ist  der  Dampfdruck $p_{ges}$ 
über der Lösung gleich dem Partialdruck des Lösungsmittels $p_1$, da der Partialdruck der
gelösten Komponente 2 vernachlässigbar klein ist.  

In  Abbildung  4.1  sind  die  Gleichgewichtskurven  für  die  Phasenübergänge  des  reinen
Lösungsmittels und die der oben beschriebenen binären Mischung in einem p-T-
Diagramm schematisch dargestellt. Die Kurven A−B, B−F und B−C stellen die Kurven des
Dampfdrucks, des Sublimationsdrucks sowie des Schmelzdrucks des reinen
Lösungsmittels  dar.  Der  Tripelpunkt  (alle  drei  Phasen  im  Gleichgewicht)  des  reinen
Lösungsmittels liegt bei B. Die Kurve D−E−F ist die Dampfdruckkurve der binären Lösung,
die Kurve F−G stellt die Schmelzdruckkurve dar und F ist deren Tripelpunkt. Wir setzen
auch  hier  wieder  voraus,  dass  die  Gasphase  aus  dem  Dampf  des  reinen  Lösungsmittels
besteht.

<div style="text-align:center;">
    <img src="Phasendiagramm.png" width=600>
    <p><b>Abbildung 4.1</b> Phasendiagramm</p>
</div>

Der  normale  Gefrierpunkt  ist  jeweils  der  Punkt  auf  der  Schmelzdruckkurve  bei  einem
Druck von $p = 101325$ Pa (1 atm) 
Der Gefrierpunkt des reinen Stoffes ist in Abbildung 4.1
demnach C und der Gefrierpunkt der Lösung ist G. Für die Gefrierpunktserniedrigung gilt
daher:
$$
    \Delta T_g = T_C - T_G \tag{1.3}
$$
Da die Schmelzdruckkurven sehr steil verlaufen, gilt außerdem $\Delta T_g \approx T_B - T_F$, das heißt, die Gefrierpunktserniedrigung ist nahezu gleich der
Erniedrigung des Tripelpunktes. Im Folgenden soll über geometrische Überlegungen aus
dem p-T-Diagramm  (Abbildung  4.1)  eine  Gleichung  für  die  Gefrierpunktserniedrigung
abgeleitet  werden.  Da  im  Versuch  nur  kleine  Gefrierpunktserniedrigungen  im  Bereich
einiger Kelvin gemessen werden, wird für die Sublimationsdruck- und Dampfdruckkurve
ein linearer Verlauf angenommen. Für die Steigung der Dampfdruckkurve der Lösung gilt:
$$
    \left.\frac{\Delta p}{\Delta T}\right|_{\text{Dampf}}=\frac{p_E - p_F}{\Delta T_g} \tag{1.4}
$$
Für die Steigung der Sublimationsdruckkurve gilt:
$$
    \left.\frac{\Delta p}{\Delta T}\right|_{\text{Subl}}=\frac{p_B - p_F}{\Delta T_g} \tag{1.5}
$$
Die Steigung sind Materialkonstanten.
Die Druckdifferenz $p_B - p_E$ entspricht der Dampfdruckerniedrigung bei der Temperatur $T_B$:
$$
    \frac{p_B - p_E}{\Delta T_g} = \left.\frac{\Delta p}{\Delta T}\right|_{\text{Subl}} - \left.\frac{\Delta p}{\Delta T}\right|_{\text{Dampf}} \tag{1.6}
$$
Nach dem Roultschem Gesetz (Gleichung (1.1)) gilt:
$$
    p_B - p_E = p_B \cdot x_2^l \tag{1.7}
$$
Einsetzen in Gleichung (1.6) führt zu folgendem Ausdruck für die Gefrierpunktserniedrigung 
$$
    \begin{align*}
        \frac{p_B \cdot x_2^l}{\Delta T_g} &= \left.\frac{\Delta p}{\Delta T}\right|_{\text{Subl}} - \left.\frac{\Delta p}{\Delta T}\right|_{\text{Dampf}} \\ \tag{1.8}
        \Delta T_g &= \frac{p_{01} \cdot x_2^l }{\left.\frac{\Delta p}{\Delta T}\right|_{\text{Subl}} - \left.\frac{\Delta p}{\Delta T}\right|_{\text{Dampf}}} = C_g \cdot x_2^l 
    \end{align*}
$$
Die Propornalitätskonstante $C_g$ ist  nur  vom  Lösungsmittel  abhängig.  Denn  wie  in
Abbildung 4.1 dargestellt, verlaufen die Dampfdruckkurve A−B des reinen Stoffes 1 und
die Dampfdruckkurve der binären Lösung F−E−D nahezu parallel. Für kleine
Konzentrationen  ist  der  Stoffmengenanteil $x_2^l$ proportional  zur  Molalität ĉ des  gelösten
Stoffes und man erhält:
$$
    \Delta T_g = \Theta_g \cdot ĉ \tag{1.9}
$$
Die Stoffkonstante $\Theta_g$ wird  als **kryoskopische**  Konstante  bezeichnet.  Ihr  Wert  beträgt
für Wasser $1.86 K\cdot Kg \cdot mol^{-1}$, für Cyclohexan $20.2 K\cdot Kg \cdot mol^{-1}$  

Die **Molalität** ĉ ist  als  der  Quotient  der  Stoffmenge  des  Gelösten  und  der  Masse  des
Lösungsmittels definiert:
$$
    ĉ = \frac{n_2}{m_1} = \frac{m_2}{M_2 \cdot m_1} \tag{1.10}
$$
Für die Gefrierpunktserniedrigung $\Delta T_g$ ergibt sich somit:
$$
    \Delta T_g = \Theta_g \cdot \frac{m_2}{M_2\cdot m_1} \tag{1.11}
$$
Offenbar  ist  die  Gefrierpunktserniedrigung  nur  von  der  Zahl  (Stoffmenge)  der  gelösten
Teilchen, nicht aber von ihrer Art abhängig. Derartige Stoffeigenschaften bezeichnet man
auch als **kolligative Eigenschaften**, zu denen beispielsweise auch die
Siedepunktserhöhung oder der osmotische Druck gehören. Äquimolare Mengen gelöster
Teilchen bedingen stets dieselbe Gefrierpunktserniedrigung. Zu beachten ist allerdings, dass die Anzahl der gelösten Teilchen nicht unbedingt mit der
Anzahl der Moleküle des gelösten Stoffes übereinstimmt. Für dissoziierende Stoffe
(beispielsweise  Salze  in  Wasser)  ist  der  Stoffmengenanteil  größer  als  er  sich  aus  der
Molmasse des undissoziierten Stoffes ergibt. Ist diese Dissoziation vollständig und führt
sie zu i Ionen pro undissoziiertem Molekül, so ist er gerade um den Faktor i größer. Für
NaCl  ist $i=2$. Ist  diese  Dissoziation  unvollständig,  so  ist  auch  der  Dissoziationsgrad  zu
berücksichtigen. Andererseits gibt es auch den Effekt der Aggregation von Molekülen in
Lösung, bei der die effektive Teilchenzahl verringert wird.

### 1.3 Siedepunkterhöhung 

Der Verlauf des Dampfdruckes $p_1$ der binären Mischung lässt sich mit Hilfe von Gleichung
(4.1)  aus  dem  Verlauf  des  Dampfdruckes $p_{01}$ der  reinen  Komponente  1  berechnen,
solange das Raoultsche Gesetz  gilt. Als Siedetemperatur wird für  eine Mischung ebenso
wie  für  einen  reinen  Stoff  die  Temperatur  bezeichnet,  bei  der  der  Gesamtdampfdruck $p_{gesamt} = 101325$ Pa beträgt.
Da in unserem Falle $p_{gesamt} = p_1 1$ ist,  entnimmt man
Abbildung 4.2, dass die Lösung den Wert 1 atm erst bei der Temperatur $T_D$ erreicht, die
reine Komponente 1 aber bereits bei der Temperatur $T_A$.

<div style="text-align:center;">
    <img src="Siedekurven.png" width=600>
    <p><b>Abbildung 4.2</b> Siedekurven des reinen Lösungsmittels und der binären Mischung</p>
</div>

Den Zusammenhang zwischen der Siedepunktserhöhung
$\Delta T_S = T_D - T_A$ und der Dampfdruckerniedrigung erhalten wir mit Hilfe von Abbildung 4.2,
die einen vergrößerten Ausschnitt der Dampfdruckkurven der reinen Komponente 1 und
der binären Mischung wiedergibt. Dabei  gilt für  die Steigung  der Dampfdruckkurve der
Lösung an irgendeinem Punkt zwischen $T_D$ und $T_A$:
$$
    \left.\frac{\Delta p}{\Delta T}\right|_{\text{Dampf}}=\frac{p_{01} - p_1}{\Delta T_S} \tag{1.12}
$$
Da $\Delta T_S$ klein  ist,  die  Steigungen  der  Dampfdruckkurve  von  Lösung  und  reinem  Stoff
ähnlich  sind  und  nicht  stark  von  der  Temperatur  abhängen,  kann  man  für  die  Steigung
der Dampfdruckkurve der Lösung in guter Näherung die Steigung der Dampfdruckkurve
des reinen Lösungsmittels am Siedepunkt $T_A$ einsetzen.  

Mit Hilfe des Raoultschen Gesetzes in Form von Gleichung (1.2)  erhält man:
$$
    \Delta T_S = \frac{p_{01}}{\left.\frac{\Delta p}{\Delta T}\right|_{Dampf}} \cdot x_2^l = C_S \cdot x_2^l \tag{1.13}
$$
Der Proportionalitätsfaktor enthält nur noch Eigenschaften des Lösungsmittels und ist für
alle Lösungen im gleichen Lösungsmittel gleich.  

Analog zu Gleichung (1.11) lässt sich eine Gleichung bilden, in dem man den
Stoffmengenanteil durch die Molalität ĉ ausdrückt:
$$
    \Delta T_S = \Theta_S \cdot ĉ = \Theta_S \cdot \frac{m_2}{M_2 \cdot m_1} \tag{1.14}
$$
Die **ebullioskopische  Konstante** $\Theta_S$  ist  für  verschiedene  Lösungsmittel  tabelliert;  für
Wasser gilt $\Theta_S = 0.512 K \cdot kg \cdot mol^{-1}$

## 2 Prinzip des Experiments

Die  Molmassen  einer  in  Cyclohexan  sowie  einer  zweiten  in  Wasser  gelösten  Substanz
sollen  bestimmt  werden.  Die  Gefrierpunktserniedrigung  wird  aus  der  Differenz  der
gemessenen Gefrierpunkte des reinen Lösungsmittels und der Lösung bestimmt.  

Zur Bestimmung der Gefrierpunkte werden die Abkühlungskurven der Lösungen
(Temperatur als Funktion der Zeit) aufgenommen. Da häufig eine Unterkühlung auftritt,
sind als Gefrierpunkte in guter Näherung die Temperaturen zu verwenden, die sich durch
lineare Extrapolation des beobachteten Temperaturganges nach beendeter Unterkühlung
auf den anfänglichen, steiler abfallenden Ast der Abkühlungskurve ergeben.  

Abbildung  4.3  zeigt  verschiedene  Typen  von  Abkühlungskurven,  die  man  erhält,  wenn
man die Temperatur der Probe gegen die Zeit aufträgt. Der Wärmefluss aus der Probe in
das Kältebad ist ungefähr proportional zu deren Temperaturdifferenz. Beim Gefrieren des
reinen Lösungsmittels (a) bleibt die Temperatur so lange konstant wie beide Phasen, fest
und  flüssig,  nebeneinander  vorliegen.  Die  freiwerdende  Kristallisationsenthalpie  fließt
dann ins Kältebad ab. Im Falle einer Lösung (b) bleibt die Temperatur in diesem Bereich
nicht konstant, sondern sinkt mit einer geringeren Steigung weiter, wodurch man einen
Knick erhält. Die Ursache dafür  ist, dass  beim Ausfrieren des reinen Lösungsmittels aus
der Lösung diese immer konzentrierter wird, so dass sich ihr Gefrierpunkt ständig weiter
erniedrigt.  

Kurven (a) und (b) sind Idealfälle. In der Praxis beobachtet man meistens eine
Unterkühlung  der  Probe,  bevor  sich  die  ersten  Kristallisationskeime  gebildet  haben  (c)
und (d). Man muss dann wie in Abbildung 4.3 gezeigt entsprechend extrapolieren, um die
Erstarrungstemperatur zu erhalten.

Bei  dem  in  Wasser  dissoziierenden  Salz gelangt man  im Falle  einer  unmittelbaren

tatsächlichen Molmasse ܯ଴ଶ zusammenhängt:
$$
    M_2 = \frac{M_{02}}{1 + (z-1)\cdot \alpha} \tag{2.1}
$$

Darin  ist z  ist  die  Anzahl  der  Teilchen,  die  von  einem  Molekül  gebildet  werden, $\alpha$ der Dissoziationsgrad. Für $\alpha = 1$ (vollständige Dissoziation) und $z = 2$ wird $M_2 = \frac{1}{2} \cdot M_{02}$.

### 3.1 Beschreibung der Apparatur 

Der zur Ermittlung der Gefrierpunktserniedrigung benutzte Beckmannsche
Gefrierpunktsapparat besteht aus einem weiten Reagenzglas mit einem seitlichen, durch
einen  Korken  verschließbaren  Einfüllstutzen  zum  Einbringen  der  zu  untersuchenden
Substanz.  Zwischen  dem  inneren  Glas,  welches  zur  Aufnahme  der  Lösung  dient,  und
einem zweiten äußeren Glas befindet sich ein Luftzwischenraum. Das Glas wird in einen
gekühlten  Aluminiumblock  gestellt.  Ein  Magnetrührer  sorgt  innerhalb  des  Gefäßes  für
eine gleichmäßige Temperaturverteilung. Die Temperaturen werden mit einem
Thermoelement bestimmt.

### 3.2 Durchführung

Jede Gruppe führt die Messungen nur für ein Lösungsmittel durch. Die Ergebnisse für das
andere Lösungsmittel besorgen Sie sich bitte von der Nachbargruppe! Das
Thermoelement wird in die Gefrierpunktsapparatur eingesetzt.  

In das innere Reagenzglas werden 20 mL des Lösungsmittels einpipettiert. Anschließend
wird der Beckmannsche Gefrierpunktsapparat in den vorgekühlten Aluminiumblock ($-30 \degree$ C bei Wasser als Lösungsmittel, $-15 \degree$ C brei Cyclohexan) gestellt. Nun wird unter
ständigem Rühren die zeitliche Temperaturänderung aufgenommen. Die zeitlichen
Intervalle  sind  sinnvoll  auszuwählen!  Häufig  sinkt  die  Temperatur  ohne  Ausscheidung
fester Substanzen erheblich unter den Gefrierpunkt. Durch Rühren oder Impfen mit einer
Spur festen Lösungsmittels kann man diese Unterkühlung aufheben, wobei die
Temperatur unter Ausscheidung fester Substanz plötzlich auf den Gefrierpunkt ansteigt.
Danach  wird so lange erwärmt,  bis die gebildete feste Substanz aufgetaut ist und durch
den  seitlichen  Ansatz  wird  die  abgewogene  Masse $m_{2a}$ der  Versuchssubstanz  in  das
innere Reagenzglas gefüllt. Für den Versuch mit Wasser als Lösungsmittel werden etwa
0.20 bis 0.40 g Versuchssubstanz eingewogen. Ist Cyclohexan das Lösungsmittel, werden
etwa 0.15 g Versuchssubstanz verwendet. Nachdem sich diese vollständig aufgelöst hat,
wird  die  Bestimmung  des  Gefrierpunktes  wiederholt.  Dieser  zweite  Versuchdurchgang
wird mit einer ähnlichen Masse $m_{2b}$ wiederholt. Beide Massen sind zu notieren.  
**Thermoelement**  
Der Temperaturverlauf wird mit einem Thermoelement elektronisch aufgenommen. Bei
einem Thermoelement sind Metalldrähte aus zwei unterschiedlichen Materialien
miteinander verlötet. In der Kontaktstelle entsteht eine kleine elektrische Spannung, die
temperaturabhängig  ist;  bei  dem  Thermoelement  Typ  K  mit  der  Metallpaarung  Ni  und
NiCr  beträgt  die  Thermospannung  in  der  Nähe  von  $0 \degree$ C etwa $40  \mu V/K$ (so  dass  z.B.  bei
Raumtemperatur von $25 \degree$ C die Thermospannung den Wert von etwa 1 mV hat).  

Die Thermospannung und ihre Änderungen sind sehr klein und kleine Störungen können
die  genaue  Temperaturmessung  verfälschen.  Ein  einzelnes  Thermoelement  ist  daher
nicht  ausreichend.  Durch  Anschluss  der  beiden  Metalldrähte  an  die  beiden  Pole  eines
Messgeräts  jeweils  treten  unterschiedliche  Metallpaarungen  auf  (Ni/Cu  und  NiCr/Cu).
Dadurch entstehen wieder zusätzliche Kontaktspannungen. Allerdings sind diese an den
beiden  Polen  unterschiedlich  und  die  Spannungsmessung  wird  verfälscht.  Man  benutzt
daher noch ein weiteres Thermoelement wie in Abbildung 4.5 gezeigt. Dieses wird genau
gegenpolig angeschlossen, so dass eine Differenzspannung zwischen den beiden
Thermoelementen gemessen wird. An beiden Anschlusspolen des Messgerätes befinden
sich  dann  auch  gleiche  Paare  (z.B.  Ni/Cu),  deren  Kontaktspannungen  sich  wegen  der
Umpolung  kompensieren.  Haben  beide  Thermolementpunkte  die  gleiche  Temperatur,
misst man als Spannung 0 V (gegensinnig verschaltet). Bringt man das zweite
Thermoelement auf eine bekannte Referenz-Temperatur (z.B. Eiswasser), kann man nun
den Temperaturunterschied genau zu dieser Referenz-Temperatur messen.  

Im  Experiment  werden  die  Thermospannungen  elektronisch  verstärkt  und  mit  einem
Digitalvoltmeter  gemessen.  Die  Messwerte  werden  über  eine  Schnittstelle  in  einen  PC
eingelesen und als ASCII-Datei abgelegt. Diese Daten werden in eine Excel-Tabelle
eingelesen und unter Berücksichtigung des Verstärkungsfaktors auf die tatsächlich
vorliegenden Spannungen umgerechnet. Anschließend werden diese Thermospannungen
mittels  einer  bereits  aufgestellten  Eichbeziehung  in  Temperaturen  umgerechnet  und
gegen  die  Zeit  aufgetragen.  Lassen  Sie  sich  für  diese  Operationen  in  den  Gebrauch  der
Messanordnung und der vorbereiteten Excel-Datenblätter einweisen.

### 4.1 Messergebnisse mit Lösungsmittel Wasser

Zuerst mussten die Massen der unbekannten Substanz gemessen werden. Für eine möglichst genaue Bestimmung wurde die Masse jeweils 10 mal gemessen auf einer feinwaage. Es ergaben sich die fogenden Werte

```python
mass_h2o = StringIO("""
#Nr.  m01/g   m02/g
1     0.2198  0.3010
2     0.2196  0.3009
3     0.2195  0.3008       
4     0.2194  0.3008
5     0.2192  0.3006
6     0.2197  0.3005
7     0.2195  0.3008
8     0.2193  0.3009 
9     0.2194  0.3010
10    0.2196  0.3007
""")

i, m01_h20, m02_h20 = np.loadtxt(mass_h2o, unpack= True)

m01_mean_h20 = np.mean(m01_h20)
m02_mean_h20 = np.mean(m02_h20)

m01_std_h2o = np.std(m01_h20)
m02_std_h2o = np.std(m02_h20)

print("Masse Messergebnisse für Messung mit Lösungsmittel Wasser")
print("Mittelwert ....... m01 = %.4f g" % m01_mean_h20)
print("Standartabweichung m01 = %.4f g" % m01_std_h2o)

print("Mittelwert ....... m02 = %.4f g" % m02_mean_h20)
print("Standartabweichung m02 = %.4f g" % m02_std_h2o)
```

```python
plt.figure(figsize=(figwidth, 0.7*figwidth))

plt.subplot(2,2,1)                      
plt.plot(i,m01_h20,'-o',label='Messung')
plt.axhline(m01_mean_h20, color='C1',label='Mittelwert')
plt.xlabel('Messung')
plt.ylabel(r'$m_{01}(H_2O)$ / g')
plt.legend(loc='upper right')

plt.subplot(2,2,2)
plt.hist(m01_h20,bins=40)
plt.axvline(m01_mean_h20, color='C1',label='Mittelwert')
plt.xlabel(r'$m_{01}(H_2O)$ / g')
plt.ylabel('Häufigkeit')
plt.legend(loc="upper right")

plt.subplot(2,2,3)
plt.plot(i,m02_h20, '-o', label="Messung")
plt.axhline(m02_mean_h20, color='C1',label='Mittelwert')
plt.xlabel('Messung')
plt.ylabel(r'$m_{02}(H_2O)$ / g')
plt.legend(loc='upper right')

plt.subplot(2,2,4)
plt.hist(m02_h20,bins=40)
plt.axvline(m02_mean_h20, color='C1',label='Mittelwert')
plt.xlabel(r'$m_{01}(H_2O)$ / g')
plt.ylabel('Häufigkeit')
plt.legend(loc="upper right")

plt.suptitle("Messungen der Masse der Probesubstanz, Lösungsmittel Wasser")
plt.tight_layout()
plt.show()
```
