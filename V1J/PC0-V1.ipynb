### 3.1. Genauigkeit der Mikrometerschraube
<font color="red"> Wozu eichen wir die Mikrometerschraube, wenn wir die erhaltenen Werte garnicht verwenden? Wir machen damit die abgelesenen Längen nur von zwei weiteren Größen ($b_1^{true}$ und $b_2^{true}$) abhängig,  deren Herkunft wir nicht kennen. Die Waage auf der anderen Seite eichen wir nicht.</font>

Die Mikrometerschraube wurde an zwei Kalibrationskörpern geeicht. Diese besaßen angegebene Breiten von
$b_1^{true} = 1.0400$ mm und $b_2^{true} = 8.0000$ mm. Bei 20 Messungen wurden die folgenden Werte erhalten:

```python
data = StringIO("""
# Nr.  b1/mm   b2/mm
1      1.0425  8.0025
2      1.0425  8.0050
3      1.0425  8.0050
4      1.0425  8.0050
5      1.0425  8.0050
6      1.0425  8.0075
7      1.0425  8.0050
8      1.0425  8.0050
9      1.0425  8.0075
10     1.0425  8.0075
11     1.0450  8.0075
12     1.0425  8.0050
13     1.0425  8.0050
14     1.0450  8.0025
15     1.0425  8.0025
16     1.0450  8.0075
17     1.0425  8.0075
18     1.0425  8.0075
19     1.0425  8.0075
20     1.0425  8.0075
""")
i, b1, b2 = np.loadtxt(data, unpack=True)
```

#### Berechnung der Mittelwerte und deren Abweichungen vom Idealwert:

<font color="blue"> Wer sich dafür interessiert, ***warum*** wir aus vielen einzelnen, voneinander unabhängigen Messungen einen ***Mittelwert*** und dann noch eine ***Standardabweichung*** oder ***Varianz*** bilden und angeben, findet unter diesem **[Link]** ein Jupyter-Notebook mit Erklärungen. </font>

```python
b1_true = 1.0400           # im Skript gegeben
b2_true = 8.0000           # im Skript gegeben
b1_mean = np.mean(b1)
b2_mean = np.mean(b2)
b1_diff = b1_mean-b1_true
b2_diff = b2_mean-b2_true

print('Mittelwert b1 = %.4f mm' % b1_mean)
print('Mittelwert b2 = %.4f mm' % b2_mean)
print('Differenz  b1 = %.4f mm' % b1_diff)
print('Differenz  b2 = %.4f mm' % b2_diff)
print('Mittelwert der Differenzen: %.5f mm' % ((b1_diff+b2_diff)/2))
```

#### Visualisierung der Messungen 
(diese Zelle kann schon wärend der laufenden Messungen ausgeführt werden): <font color="red"> (und wird vorgegeben) </font>

```python
plt.figure(figsize=(figwidth, 0.7*figwidth))

plt.subplot(2,2,1)                      # Einzelmessungen von b1
plt.plot(i,b1,'-o',label='Messung')
plt.axhline(b1_mean, color='C1',label='Mittelwert')
plt.axhline(b1_true, color='C2',label='Sollwert')
plt.xlabel('Messung')
plt.ylabel(r'$b_1$ / mm')
plt.legend(loc='upper left')

plt.subplot(2,2,2)                      # Histogramm zu b1
plt.hist(b1,bins=40,range=(1.04,1.05))
plt.axvline(b1_mean, color='C1',label='Mittelwert')
plt.axvline(b1_true, color='C2',label='Sollwert')
plt.xlabel(r'$b_1$ / mm')
plt.ylabel('Häufigkeit')

plt.subplot(2,2,3)                      # Einzelmessungen von b2
plt.plot(i,b2,'-o')
plt.axhline(b2_mean, color='C1')
plt.axhline(b2_true, color='C2')
plt.xlabel('Messung')
plt.ylabel(r'$b_2$ / mm')

plt.subplot(2,2,4)                      #Histogramm zu b2
plt.hist(b2,bins=40,range=(8.00,8.01))  
plt.axvline(b2_mean, color='C1',label='Mittelwert')
plt.axvline(b2_true, color='C2',label='Sollwert')
plt.xlabel(r'$b_2$ / mm')
plt.ylabel('Häufigkeit')

plt.suptitle('Eichung der Mikrometerschraube')
plt.tight_layout()
plt.show()
```

Für den Aluminium-Probekörper wurden die Werte für die Masse $m_{Zyl}$, die Höhe $h$, den Durchmesser $d$ und die jeweiligen dazugehörenden Mittelwerte aus der folgenden Daten bestimmt:

```python
data = StringIO("""
# Nr.  mzyl/g   h/mm       d/mm
1      9.8502  23.715625  13.998125
2      9.8504  23.713125  14.000625
3      9.8502  23.715625  14.003125
4      9.8502  23.718125  13.998125
5      9.8501  23.733125  13.998125
6      9.8503  23.713125  13.993125
7      9.8504  23.718125  13.995625
8      9.8505  23.708125  14.000625
9      9.8503  23.720625  13.893125
10     9.8500  23.720625  13.985625
""")
i, mzyl_Al, h_Al, d_Al = np.loadtxt(data, unpack=True)
```

#### Berechnung der Mittelwerte und der Standardabweichungen

```python
mzyl_Al_mean = np.mean(mzyl_Al)
mzyl_Al_std  = np.std(mzyl_Al)
h_Al_mean    = np.mean(h_Al)
h_Al_std     = np.std(h_Al)
d_Al_mean    = np.mean(d_Al)
d_Al_std     = np.std(d_Al)

print('Masse Al-Zylinder ........: (%.5f +- %.5f) g' % (mzyl_Al_mean, mzyl_Al_std))
print('Hoehe Al-Zylinder ........: (%.5f +- %.5f) mm' % (h_Al_mean, h_Al_std))
print('Durchmesser Al_Zylinder ..: (%.5f +- %.5f) mm' % (d_Al_mean, d_Al_std))
```

#### Auswertung

Das Volumen des Zylinders berechnet sich aus den Mittelwerten für $d$ und $h$ zu
$$
   V_z = \pi\left(\frac{d}{2}\right)^2 h \;\approx\; 3644.0719 \; \text{mm}^3 \;.
$$
Die Messunsicherheit dieses Wertes erhält man aus Gaußscher Fehlerfortpflanzung <font color="blue"> [siehe J-Notebook zur Fehlerrechnung] </font> mit den oben ermittelten Standardabweichungen $\Delta d$ und $\Delta h$ über
$$
   \left(\frac{\Delta V_z}{V_z}\right)^2 = 
   \left(2\frac{\Delta d}{d}\right)^2 + \left(\frac{\Delta h}{h}\right)^2    
$$
zu
$$
   \Delta V_z = V_z \; \sqrt{ 
   \left(2\frac{\Delta d}{d}\right)^2 + \left(\frac{\Delta h}{h}\right)^2}    
   \; \approx \; 16.4467 \; \text{mm}^3 \;.
$$
Also ist
$$
    V_z(\text{Al}) = (3644 \pm 16) \; \text{mm}^3 \;.
$$


```python
#~~~~~~~~~~~~~~~~~~~~~
# Ausrechnen der Werte
#~~~~~~~~~~~~~~~~~~~~~
Vz_Al = np.pi*(d_Al_mean/2)**2 * h_Al_mean
print('Volumen des Aluminium Zylinders = %.4f mm^3' % Vz_Al)

Delta_Vz_Al = np.sqrt((2*d_Al_std/d_Al_mean)**2 + (h_Al_std/h_Al_mean)**2) * Vz_Al
print('Messunsicherheit des Volumens   = %.4f mm^3' % Delta_Vz_Al)
```

#### Verbindung zur mikroskopischen Dichte

Aus Röntgenstrukturuntersuchungen weiss man, dass Aluminium ein kubisch flächenzentriertes Gitter besitzt mit einer Gitterkonstanten
$$
    a = (4.049 \pm 0.0005) \cdot 10^{-7} \text{mm}^3
$$
**Bem.:** Da im Skript keine Genauigkeit dieses Wertes angegeben ist, schätzen wir diese auf eine Einheit in der dritten Nachkommastelle. Ein kubisch flächenzentriertes Gitter besitzt
$$
     N_Z = 4
$$
Atome pro Elementarzelle. Der von uns vermessene Aluminiumzylinder enthält $V_z/a^3$ Elementarzellen und damit 
$$
      n = 4\, \frac{V_z}{a^3}\; \approx \; 2.1958 \cdot 10^{23}
$$ 
Aluminiumatome. Die Unsicherheit dieses Wertes beträgt nach Gaußscher Fehlerfortpflanzung
$$
      \left(\frac{\Delta n}{n}\right)^2 = 
      \left(\frac{\Delta V_z}{V_z}\right)^2 + \left(3\,\frac{\Delta a}{a}\right)^2\;,
$$
also $\Delta n \,\approx\, 0.01\cdot 10^{23}\;.$
Somit ist die Anzahl an Aluminiumatomen in dem vermessenen Zylinder
$$
       n = (2.20\pm 0.01) \cdot 10^{23}\;.
$$
Die Masse des Zylinders dividiert durch die Anzahl der darin enthaltenen Atome ergibt die Atommasse von Aluminium
$$
       m_{Al} = \frac{m_{zyl}}{n} \;\approx\, 4.477\cdot 10^{-23}\, \text{g/Atom}
$$
mit einer Genauigkeit von
$$
      \left(\frac{\Delta m_{Al}}{m_{Al}}\right)^2 = 
      \left(\frac{\Delta m_{zyl}}{m_{zyl}}\right)^2 + \left(\frac{\Delta n}{n}\right)^2\;,
$$
also
$$
      m_{Al} = (4.47\pm0.02)\cdot 10^{-23} \,\text{g/Atom}\;.
$$
Multipliziert man die Atommasse mit der Avogadrozahl $6.02214076\cdot10^{23}$ mol$^{-1}$ , so erhält man die Molmasse von Aluminium zu
$$
      M_{Al} = (26.96\pm0.12)\, \text{g/mol}\;.
$$
Zum Vergleich: Der Literaturwert beträgt $26.98$ g/mol.

```python
#~~~~~~~~~~~~~~~~~~~~~
# Ausrechnen der Werte
#~~~~~~~~~~~~~~~~~~~~~
print('n =',          4*3644/(4.049e-7)**3) # Anzahl Al-Atome
print('Delta n =',    2.1958*np.sqrt((16/3644)**2 + (3*0.0005/4.049)**2)) # Genauigkeit des Wertes
print('m_Al =',       9.85026/2.2e23) # m_Al in g/Atom
print('Delta_m_Al =', 4.477e-23*np.sqrt((0.00014/9.85026)**2 + (0.01/2.20)**2))
print('M_Al =',       4.477e-23*6.02214076e23)
print('Delta_M_Al =', 0.02e-23*6.02214076e23)
```
